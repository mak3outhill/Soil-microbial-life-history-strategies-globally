```r
my_theme = 
    theme(plot.title = element_text(color = "black", size = 20, hjust = 0.5),
        axis.title.x = element_text(color = "black", size = 15),
        axis.title.y = element_text(color = "black", size = 15),
        axis.text.x = element_text(vjust= 0.6,size= 10),
        axis.text.y = element_text(vjust = 1, size = 10),
        legend.key.size = unit(40, "pt"),
        legend.text = element_text(color = "black", size = 15),
        legend.background = element_rect(fill = "white", color = "white"),
        legend.key = element_rect(fill = "white"))
```

#### <font color=lightblue> The relative abundances of YAS strategies for each sample were calculated based on the RPKM abundances of microbial life-history functional genes.

```r
yas_df_wider = yas_pathway_rpkm_sum %>%
    melt(id.vars = "pathway", variable.name = "sample", value.name = "RPKM") %>%
    left_join(yas_pathway_stragety, by = "pathway") %>%
    group_by(sample, Strategy) %>%
    summarise(mean = sum(RPKM)) %>% 
    pivot_wider(names_from = Strategy, values_from = mean) %>%
    column_to_rownames(var = "sample") %>%
    apply(1, function(x) x/sum(x)) %>% t %>%
    as.data.frame %>%
    rownames_to_column(var = "sample") %>%
    left_join(meta, by = "sample")

yas_df_long = yas_df_wider[.1:4] %>% melt(id.vars = "sample", variable.name = "Strategy", value.name = "value") %>%
    left_join(meta, by = "sample")
```

#### <font color=lightblue> Fig.1b

```r
yas_long %>%
    ggplot(aes( SoilType, value,color = SoilType)) +
    geom_boxplot() +
    facet_wrap(~Strategy, scales = "free") +
    scale_color_manual(values = c("#cbc163", "#4dc493")) +
    scale_fill_manual(values = c("#cbc163", "#4dc493")) +
    theme_classic() +
    my_theme +
    #coord_flip() +
    xlab("") +
    ylab("Relative abundance") +
    theme(legend.position = "none", strip.text = element_text(size = 20), strip.background = element_rect(fill = "grey90"))
```

```r
#bp to Mbp
yas_df_long$average_genome_size = yas_df_long$average_genome_size/1e06
```

```r
ggplot() +
    geom_point(data = subset(yas_df_long, Strategy == "Growth potential" ), 
    aes(average_genome_size, value, color = SoilType), alpha = 0.2, size = 3, pch = 16) +
    geom_smooth(data = subset(yas_df_long,  Strategy == "Growth potential" & SoilType == "AS"), 
    aes(average_genome_size, value, color = SoilType), method = "lm", se = F,linewidth = 2, formula = y~poly(x,2)) +
    geom_smooth(data = subset(yas_df_long, Strategy == "Growth potential" & SoilType == "NS"), 
    aes(average_genome_size, value, color = SoilType), method = "lm", se = F,linewidth = 2, formula = y~poly(x,1)) +
    facet_wrap(~Strategy) +
    theme_bw() +
    scale_color_manual(values = c("#cbc163", "#4dc493")) +
    my_theme +
    #xlab("Average genome size (Mbp)") +
    xlab("") +
    ylab("Relative abundance") +
    theme(legend.position = "none") +
    theme(legend.position = "none", strip.text = element_text(size = 20), strip.background = element_rect(fill = "grey90"))
```
